```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    """The ODE: dy/dx = y*cos(x)"""
    return y * np.cos(x)

def exact_solution(x):
    """Analytical solution for verification"""
    return np.exp(np.sin(x))

# --- Parameters ---
h = 1.0  # Large step size for demonstration
x_start, x_end = 0, 3
x = np.arange(x_start, x_end + h, h)
n_steps = len(x)
y = np.zeros(n_steps)
y[0] = 1.0 # Initial condition

print(f"{'Step':<5} | {'x':<5} | {'y_calc':<12} | {'Method'}")
print("-" * 40)
print(f"{0:<5} | {x[0]:<5.1f} | {y[0]:<12.5e} | Initial")

# --- Phase 1: Start-up using RK4 (Steps 0 to 2) ---
for i in range(3):
    k1 = h * f(x[i], y[i])
    k2 = h * f(x[i] + h/2, y[i] + k1/2)
    k3 = h * f(x[i] + h/2, y[i] + k2/2)
    k4 = h * f(x[i] + h, y[i] + k3)
    y[i+1] = y[i] + (k1 + 2*k2 + 2*k3 + k4) / 6
    print(f"{i+1:<5} | {x[i+1]:<5.1f} | {y[i+1]:<12.5e} | RK4 Startup")

# --- Phase 2: Adams-Bashforth 4th-Order (Steps 3 onwards) ---
f_vals = [f(x[j], y[j]) for j in range(4)]

for i in range(3, n_steps - 1):
    # AB4 Formula
    y[i+1] = y[i] + (h/24) * (55*f_vals[3] - 59*f_vals[2] + 37*f_vals[1] - 9*f_vals[0])
    print(f"{i+1:<5} | {x[i+1]:<5.1f} | {y[i+1]:<12.5e} | AB4 Main")

    # Update the rolling buffer
    f_vals.pop(0)
    f_vals.append(f(x[i+1], y[i+1]))

# --- Error Calculation ---
y_exact = exact_solution(x)
error = np.abs(y - y_exact)
print("-" * 40)
print(f"Max absolute error: {np.max(error):.2e}")
```

```python
import numpy as np

def f(x, y):
    return y * np.cos(x)

def exact_solution(x):
    return np.exp(np.sin(x))

# --- Parameters ---
h = 1.0  # Reduced h slightly to better see the convergence
x_start, x_end = 0, 3
x = np.arange(x_start, x_end + h, h)
n_steps = len(x)

# Arrays for both methods
y_ab = np.zeros(n_steps)
y_am = np.zeros(n_steps)
y_ab[0] = y_am[0] = 1.0

# --- Start-up using RK4 for both ---
for i in range(3):
    # RK4 step for both to ensure identical starting conditions
    for arr in [y_ab, y_am]:
        k1 = h * f(x[i], arr[i])
        k2 = h * f(x[i] + h/2, arr[i] + k1/2)
        k3 = h * f(x[i] + h/2, arr[i] + k2/2)
        k4 = h * f(x[i] + h, arr[i] + k3)
        arr[i+1] = arr[i] + (k1 + 2*k2 + 2*k3 + k4) / 6

# Header for output
print(f"{'x':<5} | {'Exact':<12} | {'AB4 (Exp)':<12} | {'AM4 (Imp)':<12}")
print("-" * 55)

# Print startup values
for i in range(4):
    print(f"{x[i]:<5.1f} | {exact_solution(x[i]):<12.7e} | {y_ab[i]:<12.7e} | {y_am[i]:<12.7e}")

# --- Main Iteration: AB4 vs AM4 ---
for i in range(3, n_steps - 1):
    # 1. Adams-Bashforth 4th Order (Explicit)
    f0, f1, f2, f3 = f(x[i-3], y_ab[i-3]), f(x[i-2], y_ab[i-2]), f(x[i-1], y_ab[i-1]), f(x[i], y_ab[i])
    y_ab[i+1] = y_ab[i] + (h/24) * (55*f3 - 59*f2 + 37*f1 - 9*f0)

    # 2. Adams-Moulton 4th Order (Predictor-Corrector)
    # Predict with AB4
    fm0, fm1, fm2, fm3 = f(x[i-2], y_am[i-2]), f(x[i-1], y_am[i-1]), f(x[i], y_am[i]), 0
    y_predict = y_am[i] + (h/24) * (55*f(x[i], y_am[i]) - 59*f(x[i-1], y_am[i-1]) + 37*f(x[i-2], y_am[i-2]) - 9*f(x[i-3], y_am[i-3]))

    # Correct with AM4
    f_future = f(x[i+1], y_predict)
    y_am[i+1] = y_am[i] + (h/24) * (9*f_future + 19*f(x[i], y_am[i]) - 5*f(x[i-1], y_am[i-1]) + f(x[i-2], y_am[i-2]))

    print(f"{x[i+1]:<5.1f} | {exact_solution(x[i+1]):<12.5e} | {y_ab[i+1]:<12.5e} | {y_am[i+1]:<12.5e}")
```
